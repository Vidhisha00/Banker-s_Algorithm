```python
import math
import numpy as np
import sys

flag=1

class Exp(Exception):
    def __init__(self,value):
        self.value=value
    def __str__(self):
        return self.value

def Print2(allocation,need,max,Process):
    print("Process Id\t\tAllocated\t\t Maximum\t\t  Need")
    for i in range(len(need)):
        p = Process[i]
        av = ' '.join([str(a) for a in allocation[i]]).center(9)
        ma = ' '.join([str(a) for a in max[i]]).center(9)
        ne = ' '.join([str(a) for a in need[i]]).center(9)
        print(p.center(10) + "\t\t" + av + "\t\t" + ma + "\t\t" + ne)

allocation=[]
max=[]
available=[]
resource=[]
need=[]
process=[]

n=int(input("Enter total number of Resource Type : "))

for i in range(n):
    x=input("Enter the Name of Resource : ")
    resource.append(x)

# Allocated Resources in Allocated Matrix
print()
n=int(input("Enter the total number of Process : "))
print()
for j in range(n):
    temp = []
    p="P"+str(j)
    process.append(p)
    for i in resource:
        st="Enter the allocated instances of resource type "+i+" for Process "+str(j)+" : "
        t=int(input(st))
        temp.append(t)
    print()
    allocation.append(temp)

# Maximum Resources in Resource Matrix
for j in range(n):
    temp = []
    for i in resource:
        st="Enter the maximum instances of resource type "+i+" required by Process "+str(j)+" : "
        t=int(input(st))
        temp.append(t)
    print()
    max.append(temp)

# Available Number of Resource
for i in resource:
    st="Enter the available instances of resource type "+i+" in system : "
    t=int(input(st))
    available.append(t)

# Calculating Need Matrix
for i in range(n):
    a1=np.array(allocation[i])
    a2=np.array(max[i])
    t=a2-a1
    t=t.tolist()
    y=[]
    for i in t:
        try:
            if i >= 0:
                y.append(i)
            else:
                raise Exp("Invalid values of max and allocation enter!")
                print("Re-start the excution!")
        except Exp as e:
            print(e.__str__())
    need.append(y)

print()
print("Current Status of System : ")
print("Process Id\t\tAllocated\t\t Maximum\t\t  Need")
for i in range(n):
    av = ' '.join([str(a) for a in allocation[i]]).center(9)
    ma=' '.join([str(a) for a in max[i]]).center(9)
    ne=' '.join([str(a) for a in need[i]]).center(9)
    print(process[i].center(9)+"\t\t"+av+"\t\t"+ma+"\t\t"+ne)

# Determining Safe Sequence
i=0
print()
print(available)
arr1=np.array(available)
safestate=[]
while(i!=n):
    for j in range(len(need)):
        arr2=np.array(need[j])
        if(np.all(arr2<=arr1)):
            arr1=arr1+np.array(allocation[j])
            print("Available : " + str(arr1))
            print()
            allocation.pop(j)
            need.pop(j)
            max.pop(j)
            print("Process "+process[j]+" need is less then the available so hence we give available resource to "+process[j])
            safestate.append(process[j])
            process.pop(j)
            if len(need)!=0:
                print("So now State System : ")
                print()
                Print2(allocation,need,max,process)
            break;
    else:
        if len(need)!=0:
            print("DeadLock Occured!")
            flag=0
            SystemExit()
    i+=1
    print()
    
if flag==0:
  print("System is in Unsafe State!")
else:
  print("System is in safe state and the safe sequence is : ")
  print(safestate)
```
